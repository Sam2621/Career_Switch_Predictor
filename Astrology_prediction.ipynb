```python
import requests
from bs4 import BeautifulSoup

names_url_part = {}
for i in segments:
    names_href = []
    for j in range(segments[i]):
        url = "http://www.astrosage.com/celebrity-horoscope/default.asp?page=" + str(j+1) + "&category=" + str(i)
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        for row in soup.find_all(class_="ui-img-container"):
            names_href.append(row.a['href'])
    names_url_part[str(i)] = names_href
```

```python
import re

# Mapping of month names to month numbers
month_mapping = {
    'January': '01',
    'February': '02',
    'March': '03',
    'April': '04',
    'May': '05',
    'June': '06',
    'July': '07',
    'August': '08',
    'September': '09',
    'October': '10',
    'November': '11',
    'December': '12'
}

# Clean "Date of Birth" column
data["Date of Birth"] = data["Date of Birth"].astype(str)

# Define a custom parsing function
def custom_parser(date_str):
    # Regular expression pattern to match day, month, year
    pattern = r'(\w+), (\w+) (\d+), (\d+)'
    match = re.match(pattern, date_str)
    
    if match:
        day = match.group(3)
        return day
    else:
        return None

# Apply the custom parser to the "Date of Birth" column
data["day"] = data["Date of Birth"].apply(lambda x: custom_parser(x) if pd.notnull(x) else None)
data["month"] = data["Date of Birth"].str.extract(r'\w+, (\w+) \d+, \d+')
data["month"] = data["month"].map(month_mapping)
data["year"] = data["Date of Birth"].str.extract(r'\w+, \w+ \d+, (\d+)')
data["hour"] = data["Time of Birth"].str.extract(r'(\d+):')
data["minute"] = data["Time of Birth"].str.extract(r':(\d+):')

# Function to convert latitude/longitude values
def convert_lat_lon(value):
    if value is None:
        return None
    
    parts = value.split()
    numeric_parts = [int(part) for part in parts if part.isnumeric()]
    if len(numeric_parts) >= 2:
        return f"{numeric_parts[0]}.{numeric_parts[1]}"
    return None

# Apply the conversion function to Latitude and Longitude columns
data["Latitude"] = data["Latitude"].apply(convert_lat_lon)
data["Longitude"] = data["Longitude"].apply(convert_lat_lon)

data
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from catboost import CatBoostClassifier
from catboost import Pool
from sklearn.model_selection import KFold
```

```python
smote = SMOTE(sampling_strategy='minority')


 
model = CatBoostClassifier(
    iterations=50,
    learning_rate=0.3,
    random_strength=0.20,
    depth=7,
    verbose= False,
    loss_function='MultiClass',
    eval_metric='Accuracy',
    leaf_estimation_method='Newton' 
)
```

```python
import itertools
two_cat = list(itertools.combinations(mydata['Segment'].unique(), 2))
print(two_cat)
matrix = []
for t_class in two_cat:
    twoclass_data = mydata[mydata['Segment'].isin(t_class)]
    
    if len(twoclass_data['Segment'].unique()) < 2:
        continue  # Skip if there's only one class in the pair
        
    y_variable = twoclass_data['Segment']
    x_variable = twoclass_data.drop(['Segment'], axis=1)
    cv = KFold(n_splits=2, random_state=42, shuffle=True)
    scores = {}
    
    for train_index, test_index in cv.split(x_variable):
        X_train, X_test, y_train, y_test = x_variable.iloc[train_index], x_variable.iloc[test_index], y_variable.iloc[train_index], y_variable.iloc[test_index]
        
        # Assuming you've defined smote somewhere before this
        X_sm, y_sm = smote.fit_resample(X_train, y_train)
        
        train_pool = Pool(data=X_sm, label=y_sm)
        test_pool = Pool(data=X_test, label=y_test)
        
        model.fit(train_pool, plot=True, eval_set=[test_pool])  # Providing eval_set as a list
        
        scores['Pair'] = t_class
        scores['Accuracy'] = model.best_score_['validation']['Accuracy']  # Use 'validation' instead of 'validation_0'
        matrix.append(scores)

matrix
```

```python
for entry in matrix:
    entry['Accuracy'] *= 100
```

```python
acc_data = pd.DataFrame(matrix)
acc_data.to_csv("accuracydata.csv", index=None)
acc_data
```

```python
# Assuming you have trained the model and have the 'matrix' list with accuracy scores

# Define the test pair for which you want to retrieve the accuracy
test_pair = ('literature', 'criminal')

# Search for the accuracy in the 'matrix' list
accuracy = None
for entry in matrix:
    if entry['Pair'] == test_pair:
        accuracy = entry['Accuracy']
        break

# Print the accuracy if found
if accuracy is not None:
    print(f"Feasibility for pair {test_pair}: {accuracy:.2f}%")
else:
    print(f"Not feasible {test_pair}")
```
